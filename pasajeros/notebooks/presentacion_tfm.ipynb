
# 📑 Presentación TFM - Análisis del Sirio
Este notebook genera resultados estadísticos, gráficos y tablas listos para exportar y utilizar en la memoria del TFM.


```

import os, json
import pandas as pd
import plotly.express as px
from sirio import preprocessing, analysis, maps

# Crear carpeta de resultados
os.makedirs("../resultados", exist_ok=True)
```

## 1. Cargar y preprocesar los datos

```

df = pd.read_excel("../data/tus_datos_sirio.xlsx", sheet_name="PASAJEROSSIRIOv2")
df_clean = preprocessing.preprocesar_datos(df)
df_clean.head()
```

## 2. Resumen general

```

resumen = {
    "demografia": analysis.analizar_demografia(df_clean),
    "supervivencia": analysis.analizar_supervivencia(df_clean),
    "pasaje": analysis.analizar_pasaje(df_clean),
    "fechas": analysis.analizar_fechas(df_clean),
}
resumen
```

### Guardar resumen en JSON

```

with open("../resultados/estadisticas.json", "w", encoding="utf-8") as f:
    json.dump(resumen, f, indent=4, ensure_ascii=False)
print("📁 Archivo generado: resultados/estadisticas.json")
```

### Guardar datos limpios en Excel

```

df_clean.to_excel("../resultados/resumen.xlsx", index=False)
print("📁 Archivo generado: resultados/resumen.xlsx")
```

## 3. Gráficos principales

```

fig = px.histogram(df_clean, x="edad", nbins=20, title="Distribución de edades")
fig.write_image("../resultados/histograma_edades.png")
fig.show()
```

```

fig = px.pie(df_clean, names="sexo", title="Distribución por sexo")
fig.write_image("../resultados/distribucion_sexo.png")
fig.show()
```

```

fig = px.bar(df_clean["pasaje"].value_counts().reset_index(),
             x="index", y="pasaje",
             labels={"index": "Pasaje", "pasaje": "Cantidad"},
             title="Distribución por clase de pasaje")
fig.write_image("../resultados/distribucion_pasaje.png")
fig.show()
```

```

m_grad = maps.generar_mapa_graduado(df_clean, columna_valor="conteo", columna_agrupacion="municipio")
m_grad.save("../resultados/mapa_graduado.html")
m_grad
```
