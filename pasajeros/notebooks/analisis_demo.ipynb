
# 📊 Análisis de los pasajeros del Sirio
Este notebook demuestra cómo usar la librería **sirio** para analizar y visualizar datos de los pasajeros.


```

import pandas as pd
from sirio import preprocessing, analysis, maps
import plotly.express as px
```

## 1. Cargar y preprocesar los datos

```

# Cargar Excel desde la carpeta data
df = pd.read_excel("../data/tus_datos_sirio.xlsx", sheet_name="PASAJEROSSIRIOv2")

# Preprocesar (normalización de columnas y tipos)
df_clean = preprocessing.preprocesar_datos(df)

df_clean.head()
```

## 2. Estadísticas demográficas

```

demografia = analysis.analizar_demografia(df_clean)
demografia
```

## 3. Análisis de supervivencia

```

supervivencia = analysis.analizar_supervivencia(df_clean)
supervivencia
```

### Histograma de edades

```

fig = px.histogram(df_clean, x="edad", nbins=20, title="Distribución de edades")
fig.show()
```

### Distribución por sexo

```

fig = px.pie(df_clean, names="sexo", title="Distribución de pasajeros por sexo")
fig.show()
```

## 4. Distribución geográfica

```

geo = analysis.analizar_geografia(df_clean, columna="municipio")
geo
```

## 5. Distribución por pasaje

```

pasaje = analysis.analizar_pasaje(df_clean)
pasaje
```

## 6. Validación de un pasajero

```

# Selecciona un pasajero de ejemplo
pasajero = df_clean[df_clean["nombre"] == df_clean.iloc[0]["nombre"]].head(1)
resultado = analysis.validar_datos_pasajero(pasajero)
resultado
```
