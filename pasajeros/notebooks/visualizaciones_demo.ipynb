
# 🌍 Visualizaciones interactivas - Pasajeros del Sirio
Este notebook muestra ejemplos de mapas, grafos y timelines generados con el módulo `maps.py` de la librería **sirio**.


```

import pandas as pd
from sirio import preprocessing, maps
```

## 1. Cargar y preprocesar los datos

```

df = pd.read_excel("../data/tus_datos_sirio.xlsx", sheet_name="PASAJEROSSIRIOv2")
df_clean = preprocessing.preprocesar_datos(df)
df_clean.head()
```

## 2. Mapa de calor por municipio

```

m_heatmap = maps.generar_heatmap(df_clean, columna_agrupacion="municipio")
m_heatmap
```

## 3. Mapa con clusters de pasajeros

```

m_cluster = maps.generar_mapa_cluster(df_clean, columna_agrupacion="municipio")
m_cluster
```

## 4. Mapa con círculos graduados

```

m_grad = maps.generar_mapa_graduado(df_clean, columna_valor="conteo", columna_agrupacion="municipio")
m_grad
```

## 5. Mapa de rutas (embarque → hospedaje → destino)

```

m_rutas = maps.generar_mapa_rutas(df_clean)
m_rutas
```

## 6. Grafo interactivo de trayectorias

```

net = maps.generar_grafo(df_clean)
net.show("grafo_pasajeros.html")
```

## 7. Timeline de un pasajero

```

# Seleccionar pasajero de ejemplo
pasajero = df_clean[df_clean["nombre"] == df_clean.iloc[0]["nombre"]].head(1)
timeline = maps.generar_timeline(pasajero)
timeline
```
